## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
def calculate_subtotal(row):
    return row['unit_price'] * row['qty']

clients_df['subtotal'] = clients_df.apply(calculate_subtotal, axis=1)
print(clients_df[['client_id', 'unit_price', 'qty', 'subtotal']].head())





```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
def calculate_shipping_price(row):
    if row['unit_weight'] * row['qty'] > 50:
        return row['unit_weight'] * row['qty'] * 7
    else:
        return row['unit_weight'] * row['qty'] * 10
    
clients_df['shipping_price'] = clients_df.apply(calculate_shipping_price, axis=1)
clients_df[['client_id', 'unit_price', 'unit_weight', 'qty', 'shipping_price']].head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
def calculate_total_price(row):
    return (row['subtotal'] + row['shipping_price']) * 1.0925

clients_df['total_price'] = clients_df.apply(calculate_total_price, axis=1)
clients_df[['client_id', 'unit_price', 'qty', 'subtotal', 'shipping_price', 'total_price']].head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
def calculate_line_cost(row):
    return (row['unit_cost'] * row['qty']) + row['shipping_price']
                
clients_df['line_cost'] = clients_df.apply(calculate_line_cost, axis=1)
clients_df[['client_id', 'unit_price', 'unit_weight', 'qty', 'shipping_price', 'line_cost']].head()
```

```python
# Create a column for the profit of each line using line cost and line price
def calculate_line_profit(row):
    return row['total_price'] - row['line_cost']

clients_df['line_profit'] = clients_df.apply(calculate_line_profit, axis=1)
clients_df[['client_id', 'unit_price', 'unit_weight', 'qty', 'shipping_price', 'line_cost', 'line_profit']].head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
for client in top_clients_list:
    total_value = clients_df.loc[clients_df['client_id'] == client, 'total_price'].sum()
    print (f'Total value for client {client} is: {total_value:.2f}')
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
top_five_clients = {"client_id": top_clients_list}
columns_list = ['qty', 'shipping_price', 'total_price', 'line_cost', 'line_profit']
for column in columns_list:
       top_five_clients[column]= []    
for client in top_clients_list:
        for column in columns_list:
                total_value = clients_df.loc[clients_df['client_id'] == client, column].sum()
                top_five_clients[column].append(total_value)
                
summary_df = pd.DataFrame(top_five_clients)
summary_df

```

```python
# Format the data and rename the columns to names suitable for presentation.

# Define the money columns. 

# Define a function that converts a dollar amount to millions.

# Apply the currency_format_millions function to only the money columns. 

# Rename the columns to reflect the change in the money format. 

columns = ['shipping_price', 'total_price', 'line_cost', 'line_profit']
for column in columns:
    summary_df[column] = summary_df[column] / 1000000

renamed_df = summary_df.rename(columns = {
    "shipping_price": "Shipping Price (millions)",
    "total_price": "Total Revenue (millions)",
    "line_cost": "Total Cost (millions)", 
    "line_profit": "Total Profit (millions)"
})      

renamed_df.head()

```
